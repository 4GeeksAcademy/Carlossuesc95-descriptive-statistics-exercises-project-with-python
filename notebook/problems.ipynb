# Descriptive Statistics Problems

## Exercise 1 – How long are movie titles?

In this exercise, you will apply descriptive statistics techniques to a real dataset: a database of 1000 popular movies from the IMDb website.

From the title of each movie, you will:
- Calculate the title length (in characters).
- Obtain key statistical metrics: mean, median, mode, range, variance, standard deviation, skewness, and kurtosis.
- Identify the longest and shortest titles.
- Visualize the distribution of these lengths with a histogram and a density curve.


```python
# Import libraries
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Load IMDb dataset
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Create a new column called 'title_length' with the length (number of characters) of each title
df['title_length'] = df['title'].apply(len)
df['title_length']

```

```python
# Calculate mean, median, and mode of the 'title_length' column
mean_t_length = df['title_length'].mean()
median_t_length = df['title_length'].median()
mode_t_length = df['title_length'].mode()[0]
print( f"Mean: {mean_t_length}")
print( f"Median: {median_t_length}")
print( f"Mode: {mode_t_length}")
```

```python
# Calculate the range, variance, and standard deviation of 'title_length'
range_tl_length = df['title_length'].max() - df['title_length'].min()
print(f"Range: {range_tl_length}")

variance_tl_lengh = df['title_length'].var()
std = df['title_length'].std()
print(f"Variance: {variance_tl_lengh}")
print(f"Standard Deviation: {std}")

```

```python
# Calculate skewness and kurtosis of the 'title_length' column
skewness_tl_legth = df['title_length'].skew()
print(f"Skewness: {skewness_tl_legth}")

kurt_tl_legth = df['title_length'].kurt()
print(f"Kurtosis: {kurt_tl_legth}")
```

```python
# Print the shortest and longest title based on their length
max_length = df['title_length'].max()
min_legth = df['title_length'].min
print(max_length)
print(min_legth)
```

```python
# plot your results
plt.figure(figsize=(8, 6))
df['title_length'].hist(bins=40 ,edgecolor='black', alpha=0.8)
plt.title('Title length distribution')
plt.xlabel('Amount of characters')
plt.ylabel('Frequency')
plt.show()
```

## Exercise 2 – Manual Calculation of Standard Deviation

In this exercise, you will manually calculate the standard deviation by selecting five actual titles from the movie dataset you loaded earlier.


```python
# Create a list with 5 actual values from df['title_length'], for example: [10, 13, 14, 18, 22]

random_titles_length = random.choices(df['title_length'].tolist(), k=5)
random_titles_length
```

```python
# Calculate the mean of the 5 values
mean_rndm_titles_length = sum(random_titles_length) / len(random_titles_length)
print(f"Media : {mean_rndm_titles_length}")
```

```python
# Calculate the squared differences with respect to the mean

squared_diffs = sum((x - mean_rndm_titles_length)**2 for x in random_titles_length)
print(f"Suma de diferencias al cuadrado: {squared_diffs}")


```

```python
# Calculate the variance (average of the squared differences) and its square root to obtain the standard deviation


varianza_manual = squared_diffs / len(random_titles_length)
desviacion_manual = varianza_manual ** 0.5

print(f"Varianza: {varianza_manual}")
print(f"Desviación estándar: {desviacion_manual}")
```
